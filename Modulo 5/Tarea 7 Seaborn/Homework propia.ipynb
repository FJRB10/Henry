```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
wine_reviews = pd.read_csv("Datasets\wine_reviews.csv")
```

```python
wine_reviews.head()
```

1. Gráfico de dispersión que relacione "price" y "points".

```python
price = wine_reviews['price']
points = wine_reviews.points

fig = plt.figure()
ax = plt.axes()

ax = sns.scatterplot(x = "price", y = "points", data = wine_reviews)

ax.grid(True)
```

2. Un histograma con las frecuencia del campo "points".

```python
fig = plt.figure(figsize = (8,6))
ax = plt.axes()



ax = sns.histplot(data = wine_reviews, x = "points", bins = 20, kde = True)
```

3. Un boxplot del "points".

```python
plt.figure(figsize = (7,8))

sns.boxplot(data = wine_reviews, y = "points")
```

4. Un histograma con las frecuencia del campo "price".

```python
plt.figure(figsize = (8,7))

sns.histplot(data = wine_reviews[wine_reviews["price"] <= 500], x = "price", bins = 50)
```

5. Un boxplot del "price".

```python
plt.figure(figsize = (8,7))

sns.boxplot(data = wine_reviews[wine_reviews["price"] <= 100], y = "price")
```

6. Presenta la relación entre "price" y "points" en un pairplot.

```python
plt.figure(figsize = (8,7))

sns.pairplot(data = wine_reviews)
```
