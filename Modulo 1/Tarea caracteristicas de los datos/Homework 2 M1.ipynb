```python
import os
archivo = open('Emisiones_CO2.csv', 'r')
```

```python
dicc_emisiones = {  'cod_pais' : [],
                    'nom_pais' : [],
                    'region' : [],
                    'anio' : [],
                    'co2' : [],
                    'co2_percapita' : []}
```

```python
archivo = open('Emisiones_CO2.csv', 'r')

lista1 = [i for i in archivo]

for i in range(1,len(lista1)):
    datos = [lista1[i].split('|')]

    dicc_emisiones['cod_pais'].append(datos[0][0])
    dicc_emisiones['nom_pais'].append(datos[0][1])
    dicc_emisiones['region'].append(datos[0][2])
    dicc_emisiones['anio'].append(int(datos[0][3]))
    dicc_emisiones['co2'].append(datos[0][4])
    dicc_emisiones['co2_percapita'].append(datos[0][5])
    
    #print(len(datos[0][5]))
    #print(datos[0][5][0:11])

#print(lista1[26])

for i in dicc_emisiones.keys():
    print(len(dicc_emisiones[i]))
```

```python
for indice, elemento in enumerate(dicc_emisiones['co2']):
    elemento = elemento.replace('.', '')
    elemento = elemento.replace(',', '.')

    if elemento != '':
        elemento = float(elemento)
    else:
        elemento = None

    dicc_emisiones['co2'][indice] = elemento
```

```python
for indice, elemento in enumerate(dicc_emisiones['co2_percapita']):
    elemento = elemento.replace('\n', '')
    elemento = elemento.replace('.', '')
    elemento = elemento.replace(',', '.')

    if elemento != '':
        elemento = float(elemento)
    else:
        elemento = None

    dicc_emisiones['co2_percapita'][indice] = elemento
```
